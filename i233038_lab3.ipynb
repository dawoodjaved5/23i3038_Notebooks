# Task 2

```python
import random

internal_model = {'a': 'unknown', 'b': 'unknown'}
current_location = 'a'
actual_environment = {'a': 'dirty', 'b': 'dirty'}

def vacuum_agent(location, status):
    internal_model[location] = status
    if status == 'dirty':
        return 'clean'
    if location == 'a':
        return 'move b'
    else:
        return 'move a'

for step in range(1, 5):
    print(f"step {step}")
    status = actual_environment[current_location]
    print(f"location: {current_location} status: {status}")
    
    action = vacuum_agent(current_location, status)
    
    if action == 'clean':
        actual_environment[current_location] = 'clean'
        print(f"action: cleaning {current_location}")
    elif action == 'move b':
        current_location = 'b'
        print("action: moving to room b")
    elif action == 'move a':
        current_location = 'a'
        print("action: moving to room a")
        
    print(f"internal model: {internal_model}")
```

# Task 3

```python
routes = {
    'route a': {'distance': 10, 'time': 15, 'risk': 1},
    'route b': {'distance': 8, 'time': 20, 'risk': 2},
    'route c': {'distance': 12, 'time': 10, 'risk': 3}
}
def compute_utility(distance, time, risk):
    return -(distance + time + (risk * 5))

scores = {}
for name, data in routes.items():
    score = compute_utility(data['distance'], data['time'], data['risk'])
    scores[name] = score
    print(f"utility score for {name}: {score}")

best_route = max(scores, key=scores.get)

print(f"selected route: {best_route}")
```
